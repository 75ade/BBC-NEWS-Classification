```python
import pandas as pd
import numpy as np
```

```python
features = pd.read_csv(r"X_train_filtered.csv")
df_labels = pd.read_csv(r"y_train.csv")

labels = df_labels['topic']
print(type(labels))
```

```python
X_test = pd.read_csv(r"X_test_filtered.csv")
y_test = pd.read_csv(r"y_test.csv")
```

```python
X_test_origin = pd.read_csv(r"X_test_origin.csv")
print(X_test_origin)
print(X_test_origin.shape)
```

-Before we apply our training set to the model, first we need to split this training set to 2 more sets of training and test which are X_train and X_val.

-After that, we will apply those 2 sets to our model:

+Random Forrest

+SVC

-After training the 2 models, we will take the very first test set which we already had after cleaning the data to test our 2 models and we will also calculate 4 types of score: accuracy, precision, recall and f1 for each of the model.

-For each of the model, we also print the text and the prediction to that corresponding text.

-Finally, we will draw the confusion matrix of both models so that we can see how the 2 models predicted.

```python
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(features, labels, train_size=0.8, random_state=42)

print(X_train.shape)
print(X_val.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score

# Train the model
model_forrest = RandomForestClassifier(
    # n_estimators=100,            # Number of trees in the forest (start with 200)
    # max_depth=10,                # Limit depth of the trees to reduce overfitting
    # min_samples_split=10,        # Minimum samples required to split a node
    # min_samples_leaf=5,          # Minimum samples required in a leaf node
    # max_features="sqrt",         # Use sqrt(number of features) for split
    # random_state=42,             # Ensure reproducibility
    # n_jobs=-1,                   # Use all CPU cores for training
    # class_weight="balanced"      # Handle class imbalance if present
)
model_forrest.fit(X_train, y_train)

y_pred_forrest = model_forrest.predict(X_val)

# Calculate 4 points: accuracy point, precision point, recall point and f1 point
accuracy = accuracy_score(y_val, y_pred_forrest)
precision = precision_score(y_val, y_pred_forrest, average="weighted")
recall = recall_score(y_val, y_pred_forrest, average="weighted")
f1 = f1_score(y_val, y_pred_forrest, average="weighted")

print(f"Accuracy score: {accuracy:.2f}")
print(f"Precision score: {precision:.2f}")
print(f"Recall score: {recall:.2f}")
print(f"F1 score: {f1:.2f}")
```

```python
# Finally, apply the test set to the model

predictions = model_forrest.predict(X_test)

probabilities = model_forrest.predict_proba(X_test)

for i, text in enumerate(X_test_origin['text']):
    print(f"Text: {text}")
    print(f"Predicted Topic: {predictions[i]}")
    print(f"Prediction Probability: {probabilities[i]}")
    print("-------")
```

```python
from sklearn import svm
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score
from sklearn.pipeline import Pipeline

# Create a pipeline to scale features and train the SVM model
model_svm = Pipeline([
    ('scaler', StandardScaler()),  # Feature scaling
    ('svm', svm.SVC(C=100.0, kernel='linear', probability=True, class_weight='balanced'))  # Adjusted C and kernel
])

# Train the model
model_svm.fit(X_train, y_train)

# Predict on validation set
y_pred_svm = model_svm.predict(X_val)

# Calculate evaluation metrics
accuracy = accuracy_score(y_val, y_pred_svm)
precision = precision_score(y_val, y_pred_svm, average="weighted")
recall = recall_score(y_val, y_pred_svm, average="weighted")
f1 = f1_score(y_val, y_pred_svm, average="weighted")

# Print the scores
print(f"Accuracy score: {accuracy:.2f}")
print(f"Precision score: {precision:.2f}")
print(f"Recall score: {recall:.2f}")
print(f"F1 score: {f1:.2f}")
```

```python
# Apply the test set to the SVC model

predictions = model_svm.predict(X_test)

probabilities = model_svm.predict_proba(X_test)

for i, text in enumerate(X_test_origin['text']):
    print(f"Text: {text}")
    print(f"Predicted Topic: {predictions[i]}")
    print(f"Prediction Probability: {probabilities[i]}")
    print("-------")
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
# Calculate and print the confusion matrix
print("Random Forrest: \n")

conf_matrix_forrest = confusion_matrix(y_val, y_pred_forrest)
print("Confusion Matrix:\n", conf_matrix_forrest)

# Print detailed classification report
class_report_forrest = classification_report(y_val, y_pred_forrest)
print("Classification Report:\n", class_report_forrest)


print("SVC: \n")

# Calculate and print the confusion matrix
conf_matrix_svc = confusion_matrix(y_val, y_pred_svm)
print("Confusion Matrix:\n", conf_matrix_svc)

# Print detailed classification report
class_report_svc = classification_report(y_val, y_pred_svm)
print("Classification Report:\n", class_report_svc)
```
